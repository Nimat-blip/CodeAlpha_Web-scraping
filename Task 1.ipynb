```python
import requests
from bs4 import BeautifulSoup

def scrape_books(page_number):
    base_url = "http://books.toscrape.com/catalogue/page-{}.html"
    url = base_url.format(page_number)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    books = []
    for item in soup.select(".product_pod"):
        title = item.h3.a["title"]
        price = item.select_one(".price_color").text
        rating = item.p["class"][1]  # rating is stored as a class name

        books.append({
            "title": title,
            "price": price,
            "rating": rating
        })
    return books

# Example usage:
books_page1 = scrape_books(1)
for book in books_page1:
    print(book)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_books(page_number):
    base_url = "http://books.toscrape.com/catalogue/page-{}.html"
    url = base_url.format(page_number)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    books = []
    for item in soup.select(".product_pod"):
        # Extract the book title (acts like a "review title" here)
        review_title = item.h3.a["title"]
        price = item.select_one(".price_color").text
        rating = item.p["class"][1]
        availability = item.select_one(".availability").text.strip()

        books.append({
            "review_title": review_title,
            "price": price,
            "rating": rating,
            "availability": availability
        })
    return books

# Example usage:
books_page1 = scrape_books(1)
for book in books_page1:
    print(book)
```

```python
import pandas as pd

books_page1 = scrape_books(1)   # your function call
df = pd.DataFrame(books_page1)  # create the DataFrame
```
